**Voy a elegir las columnas que pueden ser mas relevantes**

```python
data = f59[['sofifa_id','age','height_cm','weight_kg','nationality','club','overall',
           'potential','work_rate','preferred_foot','value_eur', 'wage_eur','shooting',
           'passing','dribbling','defending','physic']]
```

```python
data.corr()
```

```python
data.describe()
```

Mapa de calor para ver las correlaciones entre las columnas de 'data'

```python
f,ax = plt.subplots(figsize = (14,8))
sns.heatmap(data.corr(),linewidths=.7, annot = True)
plt.show()
```

Se observa alta correlacion entre las columnas:
 * wage y value: 0.86, cuanto mas valor tiene un jugador mayor es su sueldo.
 * passing y dribbling: 0.81, el jugador habilidoso es bueno en los pases y el regate.
 * height y weight: 0.76, cuanto mas alto mas pesa el jugador.
 * dribbling y shooting: 0.76, el jugador que tiene buen regate es buen tirador.
 * overal y potential: 0.73, la media y el potencial estan relacionados.

No se puede descartar ninguna columna, ya que ninguna esta por encima del 0.95

```python
data.to_csv("./output/data_corr.csv")
```
